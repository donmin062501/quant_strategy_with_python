```python
import xlrd
```

```python
#opening an excel file using open_workbook funciton
file_path = 'C:\\Users\\User\\OneDrive\\Desktop\\QuantStudy\\Quant Strategy with Python\\엑셀 데이터\\주식 책 데이터\\python_마법공식 데이터.xlsx'
wb = xlrd.open_workbook(file_path)
```

```python
#connecting with PER sheet using sheet_by_name function
per_sh = wb.sheet_by_name('PER')
```

```python
#reading each row and storing into a dictionary
per_dict = {}
for i in range(1, per_sh.nrows):
    data = per_sh.row_values(i) #reading each row
    name = data[0]
    per = data[1]
    if per > 0: #excluding negative per values
        per_dict[name] = per
```

```python
print(per_dict.items())
```

```python
import operator

sorted_per = sorted(per_dict.items(), key = operator.itemgetter(1))
#per_dict.items() returns the list with all dictionary keys with values
#key = operator.itemgetter(1) is used to sort the dictionary using the data of value

print(sorted_per)
```

```python
#ranking based on per values
per_rank = {}
for num, firm in enumerate(sorted_per):
    per_rank[firm[0]] = num + 1
    
print(per_rank)
```

```python
#create the same ranked dictionary for ROA

roa_sh = wb.sheet_by_name('ROA')

roa_dict = {}
for i in range(1, roa_sh.nrows):
    data = roa_sh.row_values(i)
    name = data[0]
    roa = data[1]
    if roa != '': #excluding empty values
        roa_dict[name] = roa
        
print(roa_dict)
```

```python
#now sorting...
sorted_roa = sorted(roa_dict.items(), key = operator.itemgetter(1), reverse = True)
#have to reverse the order at the end due to the nature of ROA values
print(sorted_roa)
```

```python
#finally ranking

roa_rank = {}
for num, firm in enumerate(sorted_roa):
    roa_rank[firm[0]] = num + 1
    
print(roa_rank)
```

```python
#adding PER rank and ROA rank and sorting in the ascending order
#Trick: check if each firm exists in both dictionaries (per_rank & roa_rank)

total_rank = {}

for name in roa_rank.keys():
    if name in per_rank.keys():
        total_rank[name] = roa_rank[name] + per_rank[name]
        
#sorting...
sorted_total = sorted(total_rank.items(), key = operator.itemgetter(1))
print(sorted_total)
```

```python
#If wanted, can rank the firms

magic_rank = {}
for num, firm in enumerate(sorted_total):
    magic_rank[firm[0]] = num + 1
print(magic_rank)
```
