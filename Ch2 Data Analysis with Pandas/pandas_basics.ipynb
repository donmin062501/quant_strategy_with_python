```python
#Learning Series function in Pandas

import pandas as pd

pd.Series([100, 500, 150])
```

```python
#Setting my own index
pd.Series([100, 500, 150], index = ['카카오', '삼성전자', '현대차'])
```

```python
#Printing values in series

#Very similar to dictionaries
series_ex1 = pd.Series([100, 500, 150], index = ['카카오', '삼성전자', '현대차'])
print(series_ex1['카카오'])
print(series_ex1['현대차'])

#BUT, there is an order just like the lists

#can index as well
print(series_ex1[1])
```

```python
#Learning DataFrame in Pandas

pd.DataFrame({'가격':[100, 500, 150], 'PER':[0.5, 1.2, 0.2], 'ROA':[1.01, 3.1, 0.97]},
            index = ['카카오', '삼성전자', '현대차'])
```

```python
#printing values from DataFrame

df_ex1 = pd.DataFrame({'가격':[100, 500, 150], 'PER':[0.5, 1.2, 0.2], 'ROA':[1.01, 3.1, 0.97]},
            index = ['카카오', '삼성전자', '현대차'])

df_ex1['가격'] #going to print the 가격 column only
```

```python
df_ex1['가격']['삼성전자'] #Always COLUMN first, then INDEX
```

```python
#Now printing the row
df_ex1.loc['카카오']
```

```python
#Example Continued

df_ex2 = pd.DataFrame({'PER':[1.1, 0.8, 0.7, 2.3, 3.9],
                       '가격':[100, 140, 155, 70, 90],
                       '거래량':[1000, 800, 890, 700, 2000]},
                     index=['a', 'b', 'c','d', 'e'])

print(df_ex2)
```

```python
#Filtering: picking data whose 가격 (price) is equal to or greater than 100

df_ex2[df_ex2['가격'] >= 100]
```

```python
#Filtering Practice: picking data whose 거래량 is less than 1000

df_ex2[df_ex2['거래량']<1000]
```

```python
#Sorting Series
#sort_values()
#in ascending order, so if I want in descending order, I have to do: sort_values(ascending=False)

df_ex2['PER'].sort_values()
df_ex2['PER'].sort_values(ascending = False)
```

```python
#Sorting DataFrame
#We have to define how we are going to sort by naming the column

df_ex2.sort_values(by='PER') #(by='PER', ascending=False)
```

```python
#.rank()
#rank smaller data as first
#.rank(ascending=False)

df_ex2['거래량'].rank()
```

```python
df_ex2['거래량'].rank(ascending=False)
```
