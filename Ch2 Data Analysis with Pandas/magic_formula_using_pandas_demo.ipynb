```python
per_data = pd.read_excel('python_마법공식 데이터.xlsx', sheet_name='PER', index_col=0)
per_data
# index_col sets a certain column as an index
# for this case, '회사명' (firm name) is the first column, so 0
```

```python
#Now, filter all those firms whose PER values are negative

filtered_per = per_data[per_data['PER']>0]
```

```python
#Finally, sorting in ascending order

sorted_per = filtered_per.sort_values(by = 'PER')
filtered_per.sort_values(by = 'PER')
```

```python
#creating a new 'PER Ranking' column in sorted_per DataFrame
#easy to combine since sharing the same index

sorted_per['PER Ranking']=sorted_per['PER'].rank()
```

```python
sorted_per
```

```python
#Same method applies to ROA sheet

roa_data = pd.read_excel('python_마법공식 데이터.xlsx', sheet_name='ROA', index_col=0)

pd.read_excel('python_마법공식 데이터.xlsx', sheet_name='ROA', index_col=0)
# Have to drop NaN later
```

```python
filtered_roa = roa_data.dropna()
filtered_roa
```

```python
#Let's rename the column to a simpler name

filtered_roa.columns = ['ROA']
filtered_roa
```

```python
sorted_roa = filtered_roa.sort_values(by = 'ROA', ascending = False)

sorted_roa['ROA Ranking'] = sorted_roa['ROA'].rank(ascending = False)

sorted_roa
```

```python
#combining two different DataFrames: sorted_per & sorted_roa

total_df = pd.merge(sorted_per, sorted_roa, how='inner', left_index=True, right_index=True)
# how='inner' ---> merging only firms that exist in BOTH dataframes
# left/right_index = True ---> setting the reference row

total_df
```

```python
#Adding PER Rank + ROA Rank, then rank in ascending order
#Finally, sort it

total_df['Total Ranking'] = (total_df['PER Ranking'] + total_df['ROA Ranking']).rank()
total_df.sort_values(by = 'Total Ranking')
```
